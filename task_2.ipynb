```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
titanic_df = pd.read_csv("/content/titanic.csv")
```

```python
print(titanic_df.head())
```

```python
print(titanic_df.isnull().sum())
```

```python
# Replace missing values in 'Age' column with the median age
a=titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
print(a)
```

```python
b=titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
print(b)
```

```python
print(titanic_df.describe())
```

```python
print(titanic_df['Survived'].value_counts(normalize=True))
```

```python
# Fill missing values in 'Age'
titanic_df["Age"].fillna("NA", inplace = True)
print(titanic_df)
```

```python
titanic_df.dropna()
# Drop 'Cabin' column since it has many missing values
```

```python
## Exploratory Data Analysis (EDA)
# Visualize the distribution of survival
sns.countplot(x='Survived', data=titanic_df)
plt.title('Distribution of Survival')
plt.show()
```

```python
#survival distribution by gender
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.title('Survival by Gender')
plt.show()
```

```python
#Survival by Passenger Class
sns.countplot(x='Survived', hue='Pclass', data=titanic_df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
#age distribution
sns.histplot(titanic_df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.histplot(titanic_df['Fare'], bins=20, kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
#pairplot
sns.pairplot(titanic_df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')
plt.title('Pairplot of Age, Fare, Pclass, and Survival')
plt.show()
```

```python
#survival by family size
titanic_df['Family_Size'] = titanic_df['SibSp'] + titanic_df['Parch']
sns.countplot(x='Family_Size', hue='Survived', data=titanic_df)
plt.title('Survival by Family Size')
plt.show()
```
